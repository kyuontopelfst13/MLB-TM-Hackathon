Create a tool that extracts fundamental Statcast metrics (e.g., pitch speed, exit velocity) from archival game videos using computer vision. 



```python
import pandas as pd
data_file ='2024-mlb-homeruns.csv'
df = pd.read_csv(data_file)

#print(df.to_string()) 

df['title'][0]
```

```python
import cv2
from matplotlib import pyplot as plt
video_source = df['video'][0]
vidcap = cv2.VideoCapture(video_source)
frameCount = 0

while (True):
 
    success, frame = vidcap.read()
 
    if success:
        cv2.imwrite(f'/Users/love_kyuhyun123/Desktop/output/frame_{frameCount}.jpg', frame)
 
    else:
        
        break
 
    frameCount = frameCount+1
 
vidcap.release()
```

```python
from PIL import Image 
import pytesseract 
Frame_folder = '/Users/love_kyuhyun123/Desktop/output'
frameCount = 1557
pytesseract.pytesseract.tesseract_cmd = r'/opt/homebrew/Cellar/tesseract/5.5.0/bin/tesseract' 
img = cv2.imread(f'{Frame_folder}/frame_{frameCount}.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
text = pytesseract.image_to_string(img)
 
# Print the text print(text)
print(text)
plt.imshow(gray)
plt.show()
```
